```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
```

```python
dataset=pd.read_csv("churn_Modelling.csv")
```

```python
dataset.head(3)
```

```python
dataset.isnull()
```

```python
dataset = dataset.drop(['RowNumber', 'CustomerId','Surname','Geography','Gender'], axis=1)
```

```python
dataset.head()
```

```python
input_data=dataset.iloc[:,:-1]
output_data=dataset.iloc[:,-1]
```

```python
ss=StandardScaler()
input_data=pd.DataFrame(ss.fit_transform(input_data),columns=input_data.columns)#scaling korte hobe
```

```python
import tensorflow
from keras.regularizers import L2
from keras.layers import Dense,BatchNormalization,Dropout
from keras.layers import BatchNormalization
from keras.models import Sequential
from keras.callbacks import EarlyStopping
from sklearn.model_selection import train_test_split
```

```python
ann=Sequential()
```

```python
input_data.shape#8 hoilo iput size .er cheye aktu kom 6 niye kaj korbo
```

```python
x_train,x_test,y_train,y_test=train_test_split(input_data,output_data,test_size=0.2,random_state=42)
```

```python
x_train.shape
```

```python
ann.add(Dense(6,input_dim=8,activation="relu",kernel_regularizer=L2(l2=0.01)))#regularization use kortese accuracy baraite
ann.add(BatchNormalization())
ann.add(Dense(4,activation="relu",kernel_regularizer=L2(l2=0.01)))
ann.add(BatchNormalization())
ann.add(Dense(2,activation="relu",kernel_regularizer=L2(l2=0.01)))
ann.add(BatchNormalization())
ann.add(Dropout(0.3))
#1st 3 ta hidden layer,ekhane normally relu use kori 
# r output layer regression or classification onujaye apply kori
ann.add(Dense(1,activation='sigmoid')) #binary form a output dibe tai sigmoid use kori
```

```python
ann.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python

ann.fit(x_train, y_train, batch_size=100, epochs=50,validation_data=(x_test,y_test),callbacks=EarlyStopping())#overfitting overcome korte use kori validation_data
#agei stop kore dei earlystopping fole accuracy bare
```

```python
train_accuracy=ann.history.history['accuracy']
test_accuracy=ann.history.history['val_accuracy']
```

```python
plt.plot([i for i in range(1,10)],train_accuracy)
plt.plot([i for i in range(1,10)],test_accuracy,c='red')
plt.show()
```

```python
prd1=ann.predict(x_train)
prd_data1=[]
for i in prd1 :
    if i[0]>0.5:
        prd_data1.append(1)
    else:
        prd_data1.append(0)
```

```python
prd=ann.predict(x_test)
prd_data=[]
for i in prd :
    if i[0]>0.5:
        prd_data.append(1)
    else:
        prd_data.append(0)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,prd_data)*100
```

```python
accuracy_score(y_train,prd_data1)*100
```
