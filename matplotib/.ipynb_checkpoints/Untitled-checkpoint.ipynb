```python
import matplotlib.pyplot as plt
x=[1,2,3,4]
y=[2,4,6,8]
plt.plot(x,y)
plt.show()
```

```python
c=["r","y","g","b"]
plt.bar(x,y,color=c,width=.5,align="center",edgecolor="m",linewidth=5,linestyle=":",alpha=0.6) #edge 
#alpha er man 0-1 er modhe hoi...
plt.xlabel("lan",fontsize=15)
plt.ylabel("No.")
plt.title("shifa")
plt.show()
```

```python
c=["r","y","g","b"]
plt.bar(x,y,color=c,width=.5,align="center",edgecolor="m",label="khan")
plt.legend()#label dile aita likhte hobe
plt.xlabel("lan",fontsize=15)
plt.ylabel("No.")
plt.title("shifa")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=["python","c","c++","java"]
y=[85,70,60,87]
z=[20,30,40,50]
p=np.arange(len(x))
p1=[]
plt.xlabel("lan",fontsize=15)
plt.ylabel("No.")
plt.title("shifa")

plt.bar(p,y,width=0.5,color="y",label="popularity")
plt.bar(p,z,width=0.5,color="r",label="popularity1")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=["python","c","c++","java"]
y=[85,70,60,87]
z=[20,30,40,50]
width=0.2
p=np.arange(len(x))

p1=[j+width for j in p]
plt.xlabel("lan",fontsize=15)
plt.ylabel("No.")
plt.title("shifa")

plt.bar(p,y,width=0.2,color="y",label="popularity")
plt.bar(p1,z,width=0.2,color="r",label="popularity1")
plt.xticks(p+width/2,x,rotation=20)#c,c++ dekhanor jnne
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=["python","c","c++","java"]
y=[85,70,60,87]
z=[20,30,40,50]

plt.xlabel("lan",fontsize=15)
plt.ylabel("No.")
plt.title("shifa")

plt.barh(p,y,color="y",label="popularity")
plt.bar(p,z,color="r",label="popularity1")

plt.legend()
plt.show()
```

# pie plot

```python
import matplotlib.pyplot as plt
y=["python","c","c++","java"]
c=["r","y","g","b"]
x=[10,20,30,40]
ex=[0.3,0.0,0.0,0.0]
plt.pie(x,labels=y,explode=ex,colors=c,autopct="%0.1f%%",shadow=True,radius=1.5,labeldistance=1.1)
plt.show()
```

```python
import matplotlib.pyplot as plt
y=["python","c","c++","java"]
c=["r","y","g","b"]
x=[10,20,30,40]
ex=[0.3,0.0,0.0,0.0]
plt.pie(x,labels=y,explode=ex,colors=c,autopct="%0.1f%%",shadow=True,radius=1.3,labeldistance=1.1,startangle=180,
       textprops={"fontsize": 15})
plt.show()
```

```python
import matplotlib.pyplot as plt
y=["python","c","c++","java"]
c=["r","y","g","b"]
x=[10,20,30,40]
ex=[0.3,0.0,0.0,0.0]
plt.pie(x,labels=y,explode=ex,colors=c,autopct="%0.1f%%",shadow=True,radius=1,labeldistance=1.1,startangle=180,
       textprops={"fontsize": 15},counterclock=False,rotatelabels=False,wedgeprops={"linewidth":5})
plt.title("khan")
plt.legend(loc=2)
plt.show()
```

```python
import matplotlib.pyplot as plt
y=["python","c","c++","java"]
c=["r","y","g","b"]
x=[10,20,30,40]
x1=[40,30,20,10]
plt.pie(x,labels=y,radius=1.5)
plt.pie(x1,radius=1,colors=c)
plt.show()
```
